### Plotting 

Plotting কে মিডলেভেল Interface হিসেবে ব্যবহার করলেও, এইটা বেশ বেসিক জিনিশপত্র বুঝতে সাহায্য করবে আমাদের । যা কিনা অন্যান্য কনসেপ্ট গুলা গলধরন করতে পাচক হিসেবে কাজ করবে :3  

```python
import numpy as np
x = np.linspace(0, 4*np.pi, 100)
y = np.sin(x)
```

```python
from bokeh.plotting import figure

plot = figure()
plot.line(x,y)
show(plot)
```

এইখানে bokeh.plotting.figure ব্যবহার করা হইছে একটা Bokeh Plotবানানোর জন্য। এর পরে এইটাতে line() ব্যবহার করে একটা sin curve লাইন আকা হইছে । 

এই line() রে বলা হয় glyph. এই সব glyph প্যারামিটার হিসেবে ডাটা(x,y) থেকে শুরু করে stylingএর জন্য argument গুলাও নেয়।  

```python

x = np.linspace(0, 4*np.pi, 100)
y = np.sin(x)

plot = figure()
plot.xaxis.axis_label = "x"
plot.yaxis.axis_label = "amplitude"

plot.line(x,y,
         line_color='blue',
         legend='sin(x)',
         line_width=2)

plot.circle(x, 2*y, 
           fill_color = 'red',
           line_color = 'black',
           fill_alpha = 0.2,
           size = 10,
           legend='2sin(x)')

plot.line(x,np.sin(2*x),
         line_color='green',
         line_dash = [10,5,2,5],
         line_width = 2,
         legend = 'sin(2x)')
```

```python
show(plot)
```

## Bokeh Charts
Charts হচ্ছে সবচেয়ে উপরের সারির interface। Charts এর কাজ আসলে, plot গুলা সহজে তৈরী করা। কিন্তু Data manipulation এর ব্যাপারে এইটা বেশ powerful। Pandas লাইব্রেরীর DataFrame এইখানে বেশ কাজে দেয় 

charts ৪ ধরণের 

   - Scatter
   - Box
   - Histogram
   - Bar

### Scatter

```python
data.head()
```

```python
from bokeh.charts import Scatter

plot = Scatter(data, x='petal_length', y='petal_width',
               color='species',
               legend='top_left', title='Flower Morphology')
show(plot)
```

যদি মার্কারটা triangle করতে চাই । তাইলে শুধু একটা প্যারামিটার ব্যবহার করতে হবে 

```python
plot = Scatter(data, x='petal_length', y='petal_width', 
               color='species', marker='triangle',
               legend='top_left', title='Flower Morphology')
show(plot)
```

### BoxPlot

মূলত , ডাটার statistical summary ৫ টা ডাইমেনশনে - minimum, first quartile, mean, third quartile এবং maximum দেখার জন্য BoxPlot ব্যবহার করা হয় । 

```python
from bokeh.charts import BoxPlot
p = BoxPlot(
    data, label='species', values='petal_width',
    xlabel='', ylabel='petal width, mm', title='Distribution of petal widths',
    color='aqua',
)
show(p)
```

### BarCharts

Dataframe এর কলাম ধরে aggregate করে আমরা BarChart এ দেখাইতে পারি । BarChart এ agg প্যারামিটারটা aggregation algorithm হিসেবে ব্যবহার করা হয় । 
নিচের aggregation algortihm গুলা করা সম্ভব - 

- sum (default)
- mean
- count
- nunique
- median
- min
- max

```python
from bokeh.sampledata.autompg import autompg
autompg.head()
```

```python
origin_str = {1:'US', 2:'Europe', 3:'Asia'}
autompg['origin'] = autompg['origin'].map(lambda x:origin_str[x])
autompg.head()
```

```python
from bokeh.charts import Bar
p = Bar(autompg, label='yr', values='mpg', agg='mean',
       stack='origin',
       title="MPG vs year stacked by origin",legend ='top_left')

show(p)
```
